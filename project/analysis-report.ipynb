```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
from sklearn.linear_model import LinearRegression
```

```python
engine = create_engine('sqlite:///../data/temperture_change.db')
temp_df = pd.read_sql_table('data', engine)
engine = create_engine('sqlite:///../data/emissions_totals.db')
emissions_df = pd.read_sql_table('data', engine)
print(f"The number of rows in the temperature data is {temp_df.shape[0]} and the number of columns is {temp_df.shape[1]}")
print(f"The number of rows in the emissions data is {emissions_df.shape[0]} and the number of columns is {emissions_df.shape[1]}")
```

```python
temp_df = temp_df[temp_df['Area'] == 'Germany']
temp_df.head()
```

```python
emissions_df = emissions_df[emissions_df['Area'] == 'Germany']
emissions_df = emissions_df[emissions_df['Element'] == 'Emissions (CO2eq) (AR5)']
emissions_df.head()
```

```python
temp_df.columns = temp_df.columns.str.strip()

# Filter out non-month rows
months_full = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
filtered_temperature_df = temp_df[temp_df['Months'].str.strip().isin(months_full)]
fig, axes = plt.subplots(3, 4, figsize=(20, 15), sharey=True)
axes = axes.flatten()
palette = sns.color_palette("husl", len(months_full))
for i, month in enumerate(filtered_temperature_df['Months'].unique()):
    month_data = filtered_temperature_df[filtered_temperature_df['Months'] == month]
    sns.lineplot(
        x=range(2000, 2024), 
        y=month_data.iloc[:, 4:].values.flatten(), 
        marker='o', 
        linestyle='-', 
        ax=axes[i],
        color=palette[i]
    )
    axes[i].set_title(f'{month.strip()} Temperature Changes (2000-2023)', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Year', fontsize=12)
    axes[i].set_ylabel('Temperature Change (°C)', fontsize=12)
    axes[i].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0, 1.3, 0.95])
plt.suptitle('Monthly Temperature Changes (2000-2023)', fontsize=16, fontweight='bold')
plt.show()
```

```python
yearly_data = temp_df.loc[:, 'Y2000':'Y2023']
annual_mean_temps = yearly_data.mean()

# Prepare data for regression model
X = np.arange(len(annual_mean_temps)).reshape(-1, 1)
y = annual_mean_temps.values
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
sns.barplot(x=annual_mean_temps.index, y=annual_mean_temps.values, color='skyblue', label='Mean Temp Change')
plt.plot(annual_mean_temps.index, y_pred, color='red', label='Regression Line', linewidth=2)
plt.title('Annual Mean Temperature Change in Germany (2000-2023)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Mean Temperature Change (°C)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
emissions_df['Mean Emissions'] = emissions_df.loc[:, 'Y2000':'Y2023'].mean(axis=1)
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
sns.barplot(x='Mean Emissions', y='Item', data=emissions_df, hue='Item', palette='coolwarm', dodge=False, legend=False)

plt.xlabel('Mean Emissions (kt CO2eq)', fontsize=14)
plt.ylabel('Sources', fontsize=14)
plt.title('Mean CO2 Emissions for Different Items in Germany (2000-2023)', fontsize=16, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
temp_data_mean = temp_df.loc[:, 'Y2000':'Y2023'].mean()
annual_temp = temp_data_mean.rename(lambda x: x.replace('Y', '')).to_frame('Mean_Temperature')
co2_data_sum = emissions_df.loc[:, 'Y2000':'Y2023'].sum()
annual_co2 = co2_data_sum.rename(lambda x: x.strip().replace('Y', '')).to_frame('Total_CO2_Emissions')

annual_data = annual_temp.merge(annual_co2, left_index=True, right_index=True)
annual_data.reset_index(inplace=True)
annual_data.rename(columns={'index': 'Year'}, inplace=True)
annual_data['Year'] = annual_data['Year'].astype(int)
correlation = annual_data[['Mean_Temperature', 'Total_CO2_Emissions']].corr()
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
sns.scatterplot(data=annual_data, x='Mean_Temperature', y='Total_CO2_Emissions', color='blue')

plt.title('Relationship between Mean Annual Temperature and Total CO2 Emissions', fontsize=16, fontweight='bold')
plt.xlabel('Mean Annual Temperature (°C)', fontsize=14)
plt.ylabel('Total CO2 Emissions (kt)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)

plt.text(x=annual_data['Mean_Temperature'].min(), 
         y=annual_data['Total_CO2_Emissions'].max(), 
         s=f'Correlation: {correlation.iloc[0, 1]:.2f}',
         fontdict=dict(color='red', size=12, weight='bold'))

plt.show()
```
